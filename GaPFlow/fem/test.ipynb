```python
# non-periodic boundary conditions and steady-state; inclined slider use case
from GaPFlow import Problem

sim = """
options:
    output: data/inclined
    write_freq: 1000
    silent: False
grid:
    Lx: 0.1
    Ly: 1.
    Nx: 100
    Ny: 1
    xE: ['D', 'N', 'N']
    xW: ['D', 'N', 'N']
    yS: ['P', 'P', 'P']
    yN: ['P', 'P', 'P']
    xE_D: 1.1853
    xW_D: 1.1853
geometry:
    type: inclined
    hmax: 6.6e-5
    hmin: 1.0e-5
    U: 50.
    V: 0.
numerics:
    solver: fem
    CFL: 0.4
    adaptive: True
    tol: 1e-6
    dt: 1e-8
    max_it: 20_000
properties:
    EOS: PL
    shear: 1.846e-5
    bulk: 0.
    P0: 101325
    rho0: 1.1853
    alpha: 0.
fem_solver:
    type: newton_alpha
    max_iter: 15
    alpha: 1.0
    equations:
        term_list: ['R11', 'R12', 'R21', 'R24']
"""

problem = Problem.from_string(sim)
problem.run()
problem.plot()
```

```python
# periodic boundary conditions and time-stepping; journal use case
from GaPFlow import Problem

sim = """
options:
    output: data/journal
    write_freq: 1
    silent: False
grid:
    dx: 1.e-5
    dy: 1.
    Nx: 100
    Ny: 1
    xE: ['P', 'P', 'P']
    xW: ['P', 'P', 'P']
    yS: ['P', 'P', 'P']
    yN: ['P', 'P', 'P']
geometry:
    type: journal
    CR: 1.e-2
    eps: 0.7
    U: 0.1
    V: 0.
numerics:
    solver: fem
    CFL: 0.25
    adaptive: 1
    tol: 1e-8
    dt: 1e-02
    max_it: 5
properties:
    shear: 0.0794
    bulk: 0.
    EOS: DH
    P0: 101325.
    rho0: 877.7007
    T0: 323.15
    C1: 3.5e10
    C2: 1.23
fem_solver:
    type: newton_alpha
    dynamic: True
    max_iter: 10
    alpha: 1.0
    R_norm_tol: 1e-04
    equations:
        term_list: ['R11', 'R12', 'R21', 'R24', 'R1T', 'R2T']
"""

problem = Problem.from_string(sim)
problem.run()
problem.plot()
```
